```python
import pandas as pd
import os
import matplotlib.pyplot as plt

for file in os.listdir():
    col1, col2, col3 = [], [], []

    if file.endswith('.txt'):
        lines = open(file, 'r').readlines()

        for i in range(0, int(len(lines)), 3):
            col1.append(float(lines[i].strip()))
            col2.append(float(lines[i + 1].strip()))
            col3.append(float(lines[i + 2].strip()))

        # Create a pandas DataFrame
        data = {
            'time': col1,
            'probe': col2,
            'chamber': col3
        }

        df = pd.DataFrame(data).set_index('time')
        df.to_csv(file.replace('.txt', '.csv'))
        print(file)
        plt.plot(df.index, df['probe'])
        plt.plot(df.index, df['chamber'], color='r')
        
        plt.show()
```
