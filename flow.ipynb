```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
flows = pd.DataFrame([
    [50, 0.6704439999999999],
    [75, 0.45],
    [100, 0.3361465],
    [125, 0.2501017],
    [150, 0.21437309999999998],
    [175, 0.21437309999999998],
    [200, 0.15875680000000003],
    ], columns=['flow', 'mm4'])
```

```python
regr = linear_model.LinearRegression()
x = flows[['flow']]
y = flows['mm4']
regr.fit(x, y)
x = pd.DataFrame([*range(0, 210,10)], columns =['flow'])
y = regr.predict(x)
x['mm4_predict'] = y
x.set_index('flow', inplace=True)
```

```python
sns.scatterplot(flows.set_index('flow'))
sns.lineplot(x[['mm4_predict']], color='red')
plt.ylabel('Average Time (s)')
plt.xlabel('Flow (mL/min)')
```
